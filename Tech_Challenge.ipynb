## Vinhos de Mesa

```python
vinhos_de_mesa = pd.read_csv('/content/ImpVinhos (1).csv',sep = ';',thousands=".",decimal=",")
vinhos_de_mesa.head()
```

```python
# definindo o index como país para filtrar as colunas de 2008 a 2023
vinhos_de_mesa = vinhos_de_mesa.set_index('País')
```

```python
# filtrando as colunas de 2008 a 2023 para as analises
colunas_15anos = vinhos_de_mesa.columns[-32:]
vinhos_de_mesa = vinhos_de_mesa[colunas_15anos]
```

### Valor Vinho de mesa

```python
# Pegando apenas as colunas impares que são referentes ao valor importado
vinhos_de_mesa_valor = vinhos_de_mesa.iloc[:,1::2]
vinhos_de_mesa_valor
```

```python
# corringindo o nome das colunas
vinhos_de_mesa_valor.columns = vinhos_de_mesa_valor.columns.str[:-2]
```

```python
#criando uma coluna com o total de valor importado de vinhos de mesa
vinhos_de_mesa_valor['Total'] = vinhos_de_mesa_valor.sum(axis = 1)
# ordenando o total importado de vinho de mesa do maior para o menor
vinhos_de_mesa_valor = vinhos_de_mesa_valor.sort_values(by = 'Total',ascending = False)
```

```python
# ordenando o total importado de vinho de mesa do maior para o menor
vinhos_de_mesa_valor = vinhos_de_mesa_valor.sort_values(by = 'Total',ascending = False)
```

```python
# Criando uma coluna com o percentual
total_vendas = vinhos_de_mesa_valor['Total'].sum()
vinhos_de_mesa_valor['percentual'] = (vinhos_de_mesa_valor['Total'] / total_vendas) * 100
vinhos_de_mesa_valor['percentual'] = vinhos_de_mesa_valor['percentual'].round(5)
```

```python
# voltando o com uma coluna de indice
vinhos_de_mesa_valor = vinhos_de_mesa_valor.reset_index()
```

```python
# exportando para o csv
vinhos_de_mesa_valor.to_csv('Valor Importado Vinho de Mesa.csv')
```

### Vinho de Mesa Quantidade

```python
# separando as colunas com a quantidade de vinho de mesa importada
vinhos_de_mesa_quantidade = vinhos_de_mesa.iloc[:,0::2]
vinhos_de_mesa_quantidade
```

```python
#criando uma coluna com o total da quantidade importado de vinhos de mesa
vinhos_de_mesa_quantidade['Total'] = vinhos_de_mesa_quantidade.sum(axis = 1)
```

```python
# ordenando o total importado de vinho de mesa do maior para o menor
vinhos_de_mesa_quantidade = vinhos_de_mesa_quantidade.sort_values(by = 'Total',ascending = False)
```

```python
# Criando uma coluna com o percentual
quantidade_total = vinhos_de_mesa_quantidade['Total'].sum()
vinhos_de_mesa_quantidade['percentual'] = (vinhos_de_mesa_quantidade['Total'] / quantidade_total) * 100
vinhos_de_mesa_quantidade['percentual'] = vinhos_de_mesa_quantidade['percentual'].round(5)
```

```python
# voltando o com uma coluna de indice
vinhos_de_mesa_quantidade = vinhos_de_mesa_quantidade.reset_index()
```

```python
vinhos_de_mesa_quantidade
```

```python
vinhos_de_mesa_quantidade.to_csv('Quantidade Importada de Vinho de Mesa.csv')
```

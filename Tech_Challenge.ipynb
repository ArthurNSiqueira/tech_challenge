## Vinhos de Mesa

```python
vinhos_de_mesa = pd.read_csv('/content/ImpVinhos (1).csv',sep = ';',thousands=".",decimal=",")
vinhos_de_mesa.head()
```

```python
# definindo o index como país para filtrar as colunas de 2008 a 2023
vinhos_de_mesa = vinhos_de_mesa.set_index('País')
```

```python
# filtrando as colunas de 2008 a 2023 para as analises
colunas_15anos = vinhos_de_mesa.columns[-32:]
vinhos_de_mesa = vinhos_de_mesa[colunas_15anos]
```

### Valor Vinho de mesa

```python
# Pegando apenas as colunas impares que são referentes ao valor importado
vinhos_de_mesa_valor = vinhos_de_mesa.iloc[:,1::2]
vinhos_de_mesa_valor
```

```python
# corringindo o nome das colunas
vinhos_de_mesa_valor.columns = vinhos_de_mesa_valor.columns.str[:-2]
```

```python
# Resetando o index
vinhos_de_mesa_valor = vinhos_de_mesa_valor.reset_index()
# Fazendo o melt da tabela
vinhos_de_mesa_valor = vinhos_de_mesa_valor.melt(id_vars=['País'],value_vars=vinhos_de_mesa_valor.columns)
# Renomeando as colunas
vinhos_de_mesa_valor.columns = ['País','Ano','Valor']
# Exibindo os valores
vinhos_de_mesa_valor.head()
```

### Vinho de Mesa Quantidade

```python
# separando as colunas com a quantidade de vinho de mesa importada
vinhos_de_mesa_quantidade = vinhos_de_mesa.iloc[:,0::2]
vinhos_de_mesa_quantidade
```

```python
# Resetando o index
vinhos_de_mesa_quantidade = vinhos_de_mesa_quantidade.reset_index()
# Fazendo o melt da tabela
vinhos_de_mesa_quantidade = vinhos_de_mesa_quantidade.melt(id_vars=['País'],value_vars=vinhos_de_mesa_quantidade.columns)
# Renomeando as colunas
vinhos_de_mesa_quantidade.columns = ['País','Ano','Quantidade']
# Exibindo os valores
vinhos_de_mesa_quantidade.head()
```

```python
vinhos_de_mesa = pd.merge(vinhos_de_mesa_quantidade,vinhos_de_mesa_valor,on =['País','Ano'])
vinhos_de_mesa.tail(20)
```

```python
vinhos_de_mesa.to_csv('Importação Vinho de Mesa.csv')
```
